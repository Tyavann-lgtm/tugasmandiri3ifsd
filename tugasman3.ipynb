```python
import matplotlib.pyplot as plt
import seaborn as sns

# Membuat histogram untuk setiap kolom numerik
df.hist(bins=30, figsize=(15, 10))
plt.tight_layout()
plt.show()

# Membuat boxplot untuk fitur 'quality'
plt.figure(figsize=(10, 6))
sns.boxplot(x='quality', y='alcohol', df=df)
plt.title('Boxplot Kualitas vs Alkohol')
plt.show()

# Membuat scatter plot antara 'alcohol' dan 'quality'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='alcohol', y='quality', df=df)
plt.title('Scatter Plot antara Alkohol dan Kualitas')
plt.show()
```

```python
# Menghitung matriks korelasi
correlation_matrix = df.corr()

# Menampilkan matriks korelasi
print(correlation_matrix)

# Visualisasi matriks korelasi dengan heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriks Korelasi')
plt.show()
```
